```python
pdb_demo = PDB('2o2q')
```

```python
@unsync
async def pipe_demo(pdb_demo):
    assg_oper_df = await pdb_demo.fetch_from_modelServer_api(
        'atoms',
        data_collection=await pdb_demo.pipe_assg_data_collection(), 
        then_func=PDB.to_assg_oper_df)
    res2eec_df = await pdb_demo.get_res2eec_df()
    ed_as_df = await pdb_demo.fetch_from_web_api('api/pdb/entry/assembly/', PDB.assembly2eec)
    eec_as_df = await pdb_demo.get_eec_as_df() # NOTE: TO BE CHECK
    return assg_oper_df, res2eec_df, ed_as_df, eec_as_df

```

```python
%time assg_oper_df, res2eec_df, ed_as_df, eec_as_df = pipe_demo(pdb_demo).result()
```

```python
%styledf pdb_demo.fetch_from_web_api('api/pdb/entry/assembly/', PDB.to_dataframe).result()[['pdb_id', 'assembly_id', 'entity_id', 'molecule_name', 'molecule_type','in_chains']]
```

```python
assg_oper_df.merge(res2eec_df,how='outer')
```

```python
new_id_dict = dict()
for entity_id in res2eec_df.entity_id.unique():
    original_id = res2eec_df.struct_asym_id[res2eec_df.entity_id.eq(entity_id)]
    new_id = ed_as_df[ed_as_df.entity_id.eq(entity_id) & ~ed_as_df.struct_asym_id_in_assembly.isin(original_id)].struct_asym_id_in_assembly.unique()
    new_id_dict[entity_id] = new_id

new_id_check_dict = [(key, chain) for key, value in new_id_dict.items() for chain in value if chain in res2eec_df.struct_asym_id.unique()]

new_id_dict
```
